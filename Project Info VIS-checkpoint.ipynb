```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
#Creating new dataframe from the original dataframe
new_df = temp[['month', 'Current_Climatology_Trend_(1971-2020)']]
new_df.head(3)
```

```python
# sort data values
new_df.columns= ['month', 'Current_Climatology_Trend_(1971-2020)']
new_df.sort_values(by = 'Current_Climatology_Trend_(1971-2020)')
```

```python
# display all statistics
new_df.describe()
```

```python
# Split the data into train and test data
close_data = new_df['Current_Climatology_Trend_(1971-2020)'].values
close_data = close_data.reshape((-1,1))

split_percent = 0.80
split = int(split_percent*len(close_data))

x_train = close_data[:split]
y_test = close_data[split:]

date_train = new_df['month'][:split]
date_test = new_df['month'][split:]

print(len(x_train))
print(len(y_test))
```

```python
# Further split train and test data
x_train, x_test, y_train, y_test = train_test_split(x_train, date_train, test_size = 0.20, random_state=1)
```

```python
# Randomforest fit the model and predict
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
model = RandomForestClassifier(random_state=1)
model.fit(x_train, y_train)
# make predictions
yhat = model.predict(x_test)
print(yhat)
# evaluate predictions
acc = accuracy_score(y_test, yhat)
print('Accuracy: %.3f' % acc)
```

```python
#Predicting the test set results and calculating the accuracy using logistics regression
from sklearn.linear_model import LogisticRegression
logreg= LogisticRegression()
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)
print (x_test) #test dataset
print (y_pred) #predicted values
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(x_test, y_test)))
```

```python
#Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report
print('Accuracy: ',metrics.accuracy_score(y_test, y_pred))
print('Recall: ',metrics.recall_score(y_test, y_pred, zero_division=1))
print('Precision:',metrics.precision_score(y_test, y_pred, zero_division=1))
print('CL Report:',metrics.classification_report(y_test, y_pred, zero_division=1))
```
