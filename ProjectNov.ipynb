```python
close_data = df['1971-2020'].values
close_data = close_data.reshape((-1,1))

split_percent = 0.80
split = int(split_percent*len(close_data))

close_train = close_data[:split]
close_test = close_data[split:]

date_train = df['month'][:split]
date_test = df['month'][split:]

print(len(close_train))
print(len(close_test))
```

```python
from sklearn.linear_model import LinearRegression
x_train, x_test, y_train, y_test = train_test_split(close_train, date_train, test_size = 0.20, random_state=1)
```

```python
plt.scatter(x_train, y_train)
plt.xlabel('Temperature F')
plt.ylabel('Month')
plt.show()
```

```python
import statsmodels.api as sm
x = sm.add_constant(x_test)
results = sm.OLS(y_test, x).fit()
results.summary()
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
print(model.coef_)
print(model.intercept_)
```

```python
predictions = model.predict(x_test)
plt.scatter(y_test, predictions)
```
